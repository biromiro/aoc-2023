```python
import numpy as np

with open('3.txt') as f:
    lines = f.readlines()
    lines = [line.strip() for line in lines]

lines
```

```python
numbers = []
for line_idx, line in enumerate(lines):
    curr_num = ''
    had_symbol = False
    for idx, char in enumerate(line):
        if char.isdigit():
            if curr_num == '':
                curr_num = char
            else:
                curr_num += char
            # check if surrounded by symbols, diagonally as well
            # a symbol is a non-digit different than '.'
            if had_symbol:
                continue
            for up_down in [-1, 0, 1]:
                for left_right in [-1, 0, 1]:
                    if up_down == 0 and left_right == 0:
                        continue
                    if idx + left_right < 0 or idx + left_right >= len(line):
                        continue
                    if line_idx + up_down < 0 or line_idx + up_down >= len(lines):
                        continue
                    if lines[line_idx + up_down][idx + left_right] == '.':
                        continue
                    if lines[line_idx + up_down][idx + left_right].isdigit():
                        continue
                    had_symbol = True
                    break
                if had_symbol:
                    break
        else:
            if curr_num != '' and had_symbol:
                numbers.append(int(curr_num))
            curr_num = ''
            had_symbol = False
    if curr_num != '' and had_symbol:
        numbers.append(int(curr_num))
    curr_num = ''
    had_symbol = False
                
sum(numbers)
```

```python
number_indices = []
for line_idx, line in enumerate(lines):
    curr_num = ''
    curr_indices = []
    for idx, char in enumerate(line):
        if char.isdigit():
            if curr_num == '':
                curr_num = char
            else:
                curr_num += char
            curr_indices.append((line_idx, idx))
        else:
            if curr_num != '':
                number_indices.append((int(curr_num), curr_indices))
            curr_num = ''
            curr_indices = []
    if curr_num != '':
        number_indices.append((int(curr_num), curr_indices))
    curr_num = ''
    curr_indices = []

number_indices
```

```python
# find the indices of every '*' in the grid
asterisk_indices = []
for line_idx, line in enumerate(lines):
    for idx, char in enumerate(line):
        if char == '*':
            asterisk_indices.append((line_idx, idx))
asterisk_indices
```

```python
gears = []
for asterisk_line_idx, asterisk_idx in asterisk_indices:
    gear = []
    # check if the gear has any number surrounding it, diagonally as well
    # using the number_indices list
    for number, number_idxs in number_indices:
        for number_line_idx, number_idx in number_idxs:
            if abs(number_line_idx - asterisk_line_idx) <= 1 and abs(number_idx - asterisk_idx) <= 1:
                gear.append(number)
                break
    
    if len(gear) == 2:
        gears.append(np.prod(gear))

sum(gears)
```
