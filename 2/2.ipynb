```python
import numpy as np

# load data, consisting of 100 games in every row


# load 2.txt
with open('2.txt') as f:
    data = f.read().splitlines()
    games = {}
    for game in data:

        game_id, rounds = game.split(':')
        game_id = int(game_id.split(' ')[-1])
        rounds = rounds.split(';')
    
        games[game_id] = []
        
        for round_ in rounds:
            rgb = np.array([0, 0, 0])
            round_ = round_.split(',')
            
            for color in round_:
                color = [val.strip() for val in color.split(' ')]
                if color[2] == 'red':
                    # retrive the number in the color array
                    rgb[0] = int(color[1])
                elif color[2] == 'green':
                    rgb[1] = int(color[1])
                elif color[2] == 'blue':
                    rgb[2] = int(color[1])

            games[game_id].append(rgb)
            
        games[game_id] = np.array(games[game_id])
    print('Games:', games)
```

```python
matching_array = [12, 13, 14]

acc = 0
for game_id, rounds in games.items():
    for round_ in rounds:
        # if any of the indices in round is bigger than the corresponding index in matching_array, break
        if any(round_[i] > matching_array[i] for i in range(len(round_))):
            print('Game', game_id, 'is not a match')
            break
    else: 
        print('Game', game_id, 'is a match')
        acc += game_id

acc
```
